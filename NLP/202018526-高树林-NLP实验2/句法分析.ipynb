# 抽取2000条记录

```python
import pandas as pd #导入pandas库
import random
import csv
from ltp import LTP
```

```python
def GetDate(path,num):
   with open(path, "r", encoding='UTF-8') as file:
    list1 = []
    for row in csv.reader(file):
        list1.append(row[5])
    del list1[0]
    a = []
    while len(a) != num:
        b = random.randint(0, len(list1))
        if b not in a:
            a.append(b)
    sorted(a)
    return [list1[i] for i in a]
sentence = GetDate('comment.csv',2000)
print(len(sentence))
```

# 对2000条记录进行句法分析,并针对某一方面的评价得到的结果

```python
ltp = LTP(path="small")
pos_dict={'SBV':'主谓关系',
'ATT':'定中关系',
'VOB':'动宾关系',
'ADV':'状中关系',
'CMP':'动补关系',
'IOB':'间宾关系',
}
comments = {}
for sentence in GetDate('comment.csv',2000):
    seg, hidden = ltp.seg([sentence])
    dep = ltp.dep(hidden)
    sent=seg[0]
    # print(seg[0])
    dep0=dep[0]
    # print(dep0)
    for w1,w2,r in dep0:
        if r=='SBV':
            # print(sent[w1-1],sent[w2-1],pos_dict[r])
            if (w2-1,w2,'ADV') in dep0:
                if ( w1-1,w1,'ATT') in dep0:
                    comment =sent[w1-1-1] +sent[w1-1]+sent[w2-1-1]+sent[w2-1]
                    if sent[w1-1-1] not in comments:
                        comments[sent[w1-1-1]] = [comment]
                    else:
                        comments[sent[w1-1-1]].append(comment)
                else:
                    comment = sent[w1-1]+sent[w2-1-1]+sent[w2-1]
                    if sent[w1-1] not in comments:
                        comments[sent[w1-1]] = [comment]
                    else:
                        comments[sent[w1-1]].append(comment)
        if r=='CMP':
            comment = sent[w2-1]+sent[w1-1]
            if sent[w2-1] not in comments:
                comments[sent[w2-1]] = [comment]
            else:
                comments[sent[w2-1]].append(comment)
            # comments.append(comment)
            # print(sent[w2-1],sent[w1-1])
            # print(comment)
d_order=sorted(comments.items(),key=lambda x:len(x[1]),reverse=True)
# print(d_order)
for i in range(20):
    print(d_order[i][0])
    print(d_order[i][1])
```

# 对2000条记录进行句法分析,并针对所有评价得到的结果

```python
ltp = LTP(path="small")
pos_dict={'SBV':'主谓关系',
'ATT':'定中关系',
'VOB':'动宾关系',
'ADV':'状中关系',
'CMP':'动补关系',
'IOB':'间宾关系',
}
comments = {}
for sentence in GetDate('comment.csv',2000):
    seg, hidden = ltp.seg([sentence])  # 二维列表
    dep = ltp.dep(hidden)  # 二维列表
    sent=seg[0]
    dep0=dep[0]
    for w1,w2,r in dep0:
        if r=='SBV':
            if (w2-1,w2,'ADV') in dep0:
                if ( w1-1,w1,'ATT') in dep0:
                    comment =sent[w1-1-1] +sent[w1-1]+sent[w2-1-1]+sent[w2-1]
                    if comment not in comments:
                        comments[comment] = 1
                    else:
                        comments[comment] += 1
                else:
                    comment = sent[w1-1]+sent[w2-1-1]+sent[w2-1]
                    if comment not in comments:
                        comments[comment] = 1
                    else:
                        comments[comment] += 1
        if r=='CMP':
            comment = sent[w2-1]+sent[w1-1]
            if comment not in comments:
                comments[comment] = 1
            else:
                comments[comment] += 1
d_order=sorted(comments.items(),key=lambda x:x[1],reverse=True)
for i in range(20):
    print(d_order[i][0])
```
